```python
import matplotlib.pyplot as plt

# Datos de Alcaldes Electos en RM
data = {
    'Año': [2021, 2016],
    'Hombres': [36, 44],
    'Mujeres': [16, 8],
    'Rango Etario [30-39]': [20, 6],
    'Rango Etario [40-49]': [16, 22],
    'Rango Etario [50-59]': [13, 13],
    'Rango Etario [60-69]': [3, 11]
}

# Crear DataFrame con los datos
import pandas as pd
df = pd.DataFrame(data)
df.set_index('Año', inplace=True)

# Graficar los datos
plt.figure(figsize=(10, 6))

bar_width = 0.2
index = df.index

for i, col in enumerate(df.columns):
    for x, y in zip(index + (i - 2) * bar_width, df[col]):
        plt.text(x, y + 0.2, str(y), ha='center', va='bottom')

    plt.bar(index + (i - 2) * bar_width, df[col], bar_width, label=col)

plt.title('Alcaldes Electos por Año y Género en RM')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(index)
plt.legend()
plt.tight_layout()

plt.show()
```
