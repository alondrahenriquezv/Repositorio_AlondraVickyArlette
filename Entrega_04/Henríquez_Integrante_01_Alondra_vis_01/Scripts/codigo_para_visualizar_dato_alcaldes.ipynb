```python
import matplotlib.pyplot as plt

# Datos
generos = ['Hombres', 'Mujeres']
cantidad_electos = [287, 59]
total_electos = 345

# Porcentajes por género
porcentajes = [(cantidad / total_electos) * 100 for cantidad in cantidad_electos]

# Colores
colores = ['blue', 'orange']

# Etiquetas con cifras y porcentajes
etiquetas = [f'{genero}\n{cantidad} ({porcentaje:.1f}%)' for genero, cantidad, porcentaje in zip(generos, cantidad_electos, porcentajes)]

# Crear un gráfico de pastel con etiquetas
plt.figure(figsize=(6, 6))
plt.pie(porcentajes, labels=etiquetas, startangle=140, colors=colores, autopct='')
plt.title(f'Distribución de Alcaldes Electos por Género en Chile (Total: {total_electos} alcaldes)')

# Mostrar el gráfico
plt.show()
```
