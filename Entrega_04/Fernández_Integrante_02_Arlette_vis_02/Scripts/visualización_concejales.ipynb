```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos proporcionados
total_concejales = 2252
hombres = 1504
mujeres = 748

# Crear un DataFrame
data = {'Género': ['Hombres', 'Mujeres'],
        'Cantidad': [hombres, mujeres]}

df = pd.DataFrame(data)

# Visualización
plt.figure(figsize=(8, 6))
plt.pie(df['Cantidad'], labels=df['Género'], autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
plt.title('Distribución de Concejales Electos en Chile (2021-2024)')
plt.show()




```
